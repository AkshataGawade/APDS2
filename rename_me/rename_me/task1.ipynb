### 1.1 Examining and loading data

```python
# Code to inspect the provided data file...
import os
path = "./data/"
path = "./data/"
dirs = os.listdir( path )
listData=[]
for i in dirs:
    if not i.startswith('.'):
        path1=path+i
        for j in os.listdir(path+i):
            path2=path1+'/'+j
            with open(path2,'r',encoding='utf-8') as f:
                lines = f.readlines()
                result = re.fullmatch(r'(.+:\s(\d+).+)', str(lines))
                result1=re.fullmatch(r'(.+Description:\s(.*)"])', str(lines))
                if result:
                    solution1=str(result.group(2))
                if result1:
                    solution2=str(result1.group(2))
                listData.append([i,solution1,solution2])
```

### 1.2 Pre-processing data
Perform the required text pre-processing steps.

```python
df = pd.DataFrame (listData, columns = ['Job_Category', 'Webindex','Description'])
df
```

```python
# code to perform the task...
def tokenized_word(sal):
    return re.findall(r"[a-zA-Z]+(?:[-'][a-zA-Z]+)?",sal)
def remove_words(sal):
    salList=sal
    return list(filter(lambda x: len(x) >2, sal))
    
#         if len(i)<2:
#             print(len(i),i)
#             a = list(filter(lambda x: x != [0, 0], a))
            
df['Description']= df['Description'].str.lower()
df['Description_tokenized']=df['Description'].apply(lambda x: tokenized_word(x))
df['Description_tokenized_lenG2']=df['Description_tokenized'].apply(lambda x: remove_words(x))
df
```
